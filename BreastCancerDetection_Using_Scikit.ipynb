```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data = load_breast_cancer()
df = pd.DataFrame(data=cancer_data.data, columns=cancer_data.feature_names)
df['diagnosis'] = cancer_data.target
df.head()
```

# we will use pca for feature extraction due to :
#### radius_mean has around corr=1 with 'perimeter_mean','area_mean','radius_worst','perimeter_worst','area_worst' so we can drop them
#### texture_mean has around corr=1 with texture_worst so we can drop it
#### radius_se has around corr=1 with 'perimeter_se','area_se' so we can drop them
#### concavity_mean has around corr=1 with 'compactness_mean','concave points_mean' so we can drop them
#### smoothness_mean 'smoothness_worst'
#### concavity_mean  'concavity_worst','concave points_worst'
#### concavity_se  'compactness_se','concave points_se','fractal_dimension_se'


```python
corr = df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

#### get features and labels

```python
features = df.drop(columns=["diagnosis"],axis=1)
labels = df[["diagnosis"]]
```

#### we will split data into train,validate and test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.4)

X_test,X_valid,y_test, y_valid = train_test_split(X_test, y_test, test_size=0.5)
```

#### the training data will be scaled because they have different ranges
#### then we will select features using pca

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

X_valid = scaler.transform(X_valid)

pca = PCA(n_components=0.95)  # Adjust the number of components according to your needs
X_train = pca.fit_transform(X_train)
X_valid = pca.transform(X_valid)
print("number of features used = ",X_train.shape[1])



```

```python
model = LogisticRegression()

clf = model.fit(X_train,y_train)
```

#### we will validate model before testing

```python
clf.score(X_valid,y_valid)
```

#### we will scale test data according to training data information then we will get accuracy

```python
X_test = scaler.transform(X_test)
X_test = pca.transform(X_test)
clf.score(X_test,y_test)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
report = classification_report(y_test, y_pred)
print(f'Classification Report:\n{report}')
```
